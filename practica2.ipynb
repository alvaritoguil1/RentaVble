Las acciones de Tesla tienen un valor en bolsa de 999. Su variacion diaria media es del 1%, y queremos saber cual sera su cotizacion dentro de 30 dias y como ha variado respecto a la actual.   

```python
import random
import matplotlib.pyplot as plt
random.seed()
Pre_tesla = 999     # inicializamos el valor actual de Tesla
n = 30              # indicamos el numero de dias de cotizacion que queremos saber
factor = 1.01       # como su variación media es del 1%, el factor sera de 1.01
p=0.5               # la probalidad que suba la acccion un 1% es del 50%
Cot_tesla = [Pre_tesla] # inicializamos la serie de precios, comenzamos con la primera cotización
for t in range(1,n):
    if random.random() < p:
        Pre_tesla *= factor
    else:
        Pre_tesla /= factor
    Cot_tesla.append(Pre_tesla)
plt.plot(range(n),Cot_tesla)
plt.show()               #Mostramos el grafico de la cotizacion de Tesla durante los 30 dias
Var_tesla= (Pre_tesla/999) - 1
print(f"La cotizacion de tesla ha variado un {Var_tesla:.2%} ")    #dividimos el precio final de tesla entre el inicial 

                                                                  #para saber en que porcentaje ha variado su cotizacion

```

RAMDON WALK EN DOS DIMENSIONES

```python
# Ahora siguiendo el jemeplo anterior podemos ver como se mueve una accion en dos dimensiones de manera aleatoria
# Se parte del punto (0,0) del plano cartesiano  
# Cada paso que se da es de longitud r  
# El ángulo con el que se da cada paso es aleatorio  
import random  
import math  
import matplotlib.pyplot as plt  
random.seed()  
n=30 # número de pasos  
r=1 # longitud del paso  
x=y=0 # punto de partida (x,y)=(0,0)  
listax=[x]  
listay=[y]  
for i in range(n):  
  angulo=random.random()*math.pi*2 # en radianes  
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  listax.append(x)  
  listay.append(y)  
plt.plot(listax,listay)  
plt.show()  
```
