Las acciones de Tesla tienen un valor en bolsa de 999. Su variacion diaria media es del 1%, y queremos saber cual sera su cotizacion dentro de 30 dias y como ha variado respecto a la actual.   

```python
import random
import matplotlib.pyplot as plt
random.seed()
Pre_tesla = 999     # inicializamos el valor actual de Tesla
n = 30              # indicamos el numero de dias de cotizacion que queremos saber
factor = 1.01       # como su variación media es del 1%, el factor sera de 1.01
p=0.5               # la probalidad que suba la acccion un 1% es del 50%
Cot_tesla = [Pre_tesla] # inicializamos la serie de precios, comenzamos con la primera cotización
for t in range(1,n):
    if random.random() < p:
        Pre_tesla *= factor
    else:
        Pre_tesla /= factor
    Cot_tesla.append(Pre_tesla)
plt.plot(range(n),Cot_tesla)
plt.show()
Var_tesla= (Pre_tesla/999) - 1
print(f"La cotizacion de tesla ha variado un {Var_tesla:.2%} ")

```
